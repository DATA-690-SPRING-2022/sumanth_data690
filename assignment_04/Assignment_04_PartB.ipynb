```python
import pandas as pd
df = pd.read_csv('/content/census_costs.txt', delimiter = "\t",header=1)
```

```python
def cleanup(x):
  return int(x.replace(",","").replace("*",""))
```

```python
df["Census_Year_New"] = df["Census Year"].apply(cleanup)  
df["Total_Pop_New"] = df["Total Population"].apply(cleanup)
df
```

```python
def cleanup2(x):
  return (x.replace(",","").replace("$","").replace(" Billion","000000000").replace("cents",""))
```

```python
df["cost_new"] = df["Census Cost"].apply(cleanup2)
df["avg_new"] = df["Average Cost Per Person"].apply(cleanup2)
df
```

```python
df['cost_new'] = pd.to_numeric(df['cost_new'])
df['avg_new'] = pd.to_numeric(df['avg_new'])
```
